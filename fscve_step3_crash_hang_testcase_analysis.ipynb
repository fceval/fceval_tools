```python
from fscve_step1_db import AnalysisState,FscveStep1DbHelper
dbhelper = FscveStep1DbHelper( "nm", "fca",  1, False)
session = dbhelper.session
expid = dbhelper.experiment_id
benchmark = dbhelper.benchmark
fc = dbhelper.fc
query = session.query(AnalysisState.analysis_id,AnalysisState.discovery_id).first()
print(query)
```

```python
from fscve_step1_db import FuzzerType, TestCaseType, TestCase, Discovery, Fuzzer
import pandas as pd

#连接查询  哪个fuzzer（什么类型）在什么时间发现了哪个crash用例测试
querycrash =  session.query(Discovery.discovery_id,Discovery.discovery_fuzzer,Fuzzer.fuzzer_name,FuzzerType.description.label("fuzzer_type_description"),Discovery.discovery_time,TestCase.hash.label("testcase_hash"),TestCaseType.description.label("testcase_type_description"))
querycrash = querycrash.join((Fuzzer,Discovery.discovery_fuzzer==Fuzzer.fuzzer_id),(FuzzerType,Fuzzer.fuzzer_type_id==FuzzerType.id)).join((TestCase,Discovery.test_case_hash == TestCase.hash),(TestCaseType,TestCase.test_case_type_id == TestCaseType.id)).filter(TestCaseType.id==2)
querycrash = querycrash.all()
print(querycrash)
#print(querycrash.keys())
dict_value_discovery_id = [result.discovery_id  for result in querycrash]
dict_value_discovery_fuzzer = [result.discovery_fuzzer  for result in querycrash]
dict_value_discovery_fuzzer_name = [result.fuzzer_name  for result in querycrash]
dict_value_fuzzer_type_description = [result.fuzzer_type_description  for result in querycrash]
dict_value_discovery_time = [result.discovery_time  for result in querycrash]
dict_value_testcase_hash = [result.testcase_hash  for result in querycrash]
dict_value_testcase_type_description = [result.testcase_type_description  for result in querycrash]
dictcrash = {"discovery_id":dict_value_discovery_id, "discovery_fuzzer":dict_value_discovery_fuzzer, "discovery_fuzzer_name":dict_value_discovery_fuzzer_name, "fuzzer_type_description":dict_value_fuzzer_type_description, "discovery_time":dict_value_discovery_time, "testcase_hash":dict_value_testcase_hash, "testcase_type_description":dict_value_testcase_type_description}
# df_crash = pd.DataFrame(dictcrash,columns=["discovery_id", "discovery_fuzzer", "discovery_fuzzer_name", "fuzzer_type_description", "discovery_time",  "testcase_hash", "testcase_type_description"])
df_crash = pd.DataFrame(dictcrash)
print(df_crash)
path_crash_csv = f"csv/fscve_crash_testcase_{benchmark}_{fc}_{expid}.csv"
df_crash.to_csv(path_crash_csv)
```
