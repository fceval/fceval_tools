```python
#统计数量
#内置函数使用  func.sum func.count
#label函数可以为们访问的列起一个有用的名字

from sqlalchemy import func
count_case_type = session.query(func.count(TestCase.test_case_type_id)).first()
print(count_case_type)
count_case_type = session.query(func.count(TestCase.test_case_type_id)).scalar()
print(count_case_type)
count_case_type = session.query(func.count(TestCase.test_case_type_id).label('typecount')).first()
print(count_case_type)
print(count_case_type.keys())
print(count_case_type.typecount)
```

```python
#统计数量
#内置函数使用  func.sum func.count
#label函数可以为们访问的列起一个有用的名字
#过滤 filter布尔值 filter_by关键字赋值
#or_,and_,not_ 连接词表多个过滤条件
#可以直接在查询列上进行运算 99+

from sqlalchemy import func,or_
count_case_type = session.query(func.count(TestCase.test_case_type_id)).scalar()
print(count_case_type)
count_case_type = session.query(TestCase,func.count(TestCase.test_case_type_id).label("count000")).filter(TestCase.hash.startswith("000")).first()
print(count_case_type.count000)
count_case_type = session.query(TestCase,(99+func.count(TestCase.test_case_type_id)).label("count000")).filter(TestCase.hash.like("000%"),TestCase.test_case_type_id==1).first()
print(count_case_type.count000)

count_case_type = session.query(TestCase,func.count(TestCase.test_case_type_id).label("count000")).filter(or_(TestCase.hash.like("000%"),TestCase.test_case_type_id==1)).first()
print(count_case_type.count000)
```
