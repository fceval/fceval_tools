```python
import pandas as pd
expids = 10  #repeated times
benchmarks = {"cflow", "exiv2", "jq", "mp42aac"}
fcs = {"fca", "fcb"}
df = pd.DataFrame()

#calc all data of a fuzzer combination on one benchmark
for benchmark in benchmarks:
    df_results = []
    for fc in fcs:
        for expid in range(1,expids+1):
            df = pd.read_csv(f"csvzzzbugcase/{benchmark}_{fc}_{expid}.csv")
            # print(df)
            df_results.append(df)
    #merge two fuzzer combination
    df_result_final = pd.concat(df_results, ignore_index=True)
    print(df_result_final)
    df_result_final = df_result_final.drop_duplicates(['crash_frames_hash'])
    # print(df_result_final)
    if not df_result_final.empty:
        df_result_final.to_csv(f"csvzzzbugcase/{benchmark}.csv")
```
